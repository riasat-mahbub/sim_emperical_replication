```python
def remove_outliers(df, threshold=3, columns=None):
    """
    Remove outliers from DataFrame using Z-score method.
    
    Parameters:
    - df: Input pandas DataFrame.
    - threshold: Z-score threshold for outlier detection (default: 3).
    - columns: List of columns to apply outlier removal (default: all columns).
    
    Returns:
    - DataFrame with outliers removed.
    """
    # Select columns if specified, otherwise use all columns
    if columns is None:
        columns = df.columns
    
    # Calculate Z-scores for selected columns
    z_scores = df[columns].apply(lambda x: (x - x.mean()) / x.std())
    
    # Filter out rows with outliers
    filtered_df = df[(z_scores.abs() < threshold).all(axis=1)]
    
    return filtered_df
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import mannwhitneyu
from cliffs_delta import cliffs_delta

migsdf = pd.read_csv("Implementation Smell Simulation Group.csv ")
migtdf = pd.read_csv("Implementation Smell Traditional Group.csv")

migsdf = remove_outliers(migsdf,1, ['Avg'])
migtdf = remove_outliers(migtdf,1, ['Avg'])

statistic, p_value = mannwhitneyu(migsdf['Avg'], migtdf['Avg'])
plt.figure(facecolor='white')
plt.boxplot([migsdf['Avg'], migtdf['Avg']], labels=['Simulation Software', 'Traditional Software'], patch_artist=True )
plt.title("Implementation Smells")
plt.savefig('a.png')
plt.show()

print("Mann-Whitney U statistic:", statistic)
print("p-value:", p_value)

cf_delta = cliffs_delta(migsdf['Avg'], migtdf['Avg'])


if p_value < 0.05:
    print("The difference in distributions is statistically significant")
else:
    print("The difference in distributions is not statistically significant")
```
