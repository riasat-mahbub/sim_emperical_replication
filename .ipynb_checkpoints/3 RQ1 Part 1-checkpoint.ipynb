```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import mannwhitneyu
from cliffs_delta import cliffs_delta

migsdf = pd.read_csv("Implementation Smell Simulation.csv ")
migtdf = pd.read_csv("Implementation Smell Traditional.csv")

statistic, p_value = mannwhitneyu(migsdf['Avg'], migtdf['Avg'])
plt.figure(facecolor='white')
plt.boxplot([migsdf['Avg'], migtdf['Avg']], labels=['Simulation Software', 'Traditional Software'], patch_artist=True )
plt.title("Implementation Smells")
plt.savefig('a.png')
plt.show()

print("Mann-Whitney U statistic:", statistic)
print("p-value:", p_value)

cf_delta = cliffs_delta(migsdf['Avg'], migtdf['Avg'])
print(cf_delta)


if p_value < 0.05:
    print("The difference in distributions is statistically significant")
else:
    print("The difference in distributions is not statistically significant")
```

```python
import matplotlib.pyplot as plt
from scipy.stats import mannwhitneyu

mdgsdf = pd.read_csv("Design Smell Simulation.csv ")
mdgtdf = pd.read_csv("Design Smell Traditional.csv")

statistic, p_value = mannwhitneyu(mdgsdf['Avg'], mdgtdf['Avg'])

plt.boxplot([mdgsdf['Avg'], mdgtdf['Avg']], labels=['Simulation Software', 'Traditional Software'] , patch_artist=True)
plt.title("Design Smells")
plt.show()

print("Mann-Whitney U statistic:", statistic)
print("p-value:", p_value)

cf_delta = cliffs_delta(mdgsdf['Avg'], mdgtdf['Avg'])
print(cf_delta)


if p_value < 0.05:
    print("The difference in distributions is statistically significant")
else:
    print("The difference in distributions is not statistically significant")
```

```python
import matplotlib.pyplot as plt
from scipy.stats import mannwhitneyu

magsdf = pd.read_csv("Architecture Smell Simulation.csv")
magtdf = pd.read_csv("Architecture Smell Traditional.csv")

statistic, p_value = mannwhitneyu(magsdf['Avg'], magtdf['Avg'])

plt.boxplot([magsdf['Avg'], magtdf['Avg']], labels=['Simulation Software', 'Traditional Software'], patch_artist=True )
plt.title("Architectural Smells")
plt.show()

print("Mann-Whitney U statistic:", statistic)
print("p-value:", p_value)

cf_delta = cliffs_delta(magsdf['Avg'], magtdf['Avg'])
print(cf_delta)


if p_value < 0.05:
    print("The difference in distributions is statistically significant")
else:
    print("The difference in distributions is not statistically significant")
```

```python
fig, ax = plt.subplots(figsize =(15, 8))
box1 = ax.boxplot([migsdf['Avg'], mdgsdf['Avg'], magsdf['Avg']], labels=["Implementation Smell", "Design Smell", "Architecture Smell"], positions=[1, 2, 3], patch_artist=True,  boxprops=dict(facecolor="red", color="red"))
box2 = ax.boxplot([migtdf['Avg'], mdgtdf['Avg'], magtdf['Avg']],  labels=["Implementation Smell", "Design Smell", "Architecture Smell"],positions=[5, 6, 7], patch_artist=True)
plt.ylim(0,0.3)


# Create a custom legend
legend_labels = ['Simulation Software', 'Traditional Software']
legend_colors = ['red', 'blue']
custom_legend = [plt.Line2D([0], [0], marker='s', color='w', markerfacecolor=color, markersize=10, label=label)
                 for label, color in zip(legend_labels, legend_colors)]

# Add the legend to the figure
fig.legend(handles=custom_legend, loc='lower left', bbox_to_anchor=(0.75, 0.75), title="Legend")

plt.ylabel('Value')
plt.title('')
plt.savefig('Comparison.png')
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
categories = ["Implementation Smell", "Design Smell", "Architecture Smell"]
simulation_data = [0.1, 0.15, 0.2]
traditional_data = [0.05, 0.1, 0.15]

fig, ax = plt.subplots(figsize=(15, 8))

bar_width = 0.35
bar1 = ax.bar(np.arange(len(categories)), simulation_data, bar_width, label='Simulation Software', color='red')
bar2 = ax.bar(np.arange(len(categories)) + bar_width, traditional_data, bar_width, label='Traditional Software', color='blue')

ax.set_xlabel('Smell Type')
ax.set_ylabel('Value')
ax.set_title('Comparison of Software Smells between Simulation and Traditional Software')
ax.set_xticks(np.arange(len(categories)) + bar_width / 2)
ax.set_xticklabels(categories)
ax.legend()
```
